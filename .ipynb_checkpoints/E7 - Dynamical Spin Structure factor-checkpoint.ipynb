We want to compute\begin{equation}
S^{zz}(j,t_n) = \left\langle 0 | s^z_j(t_n) s^z_{L/2}(0)|0\right\rangle - \left\langle 0 | s^z_j s^z_{L/2}|0\right\rangle
\end{equation}

```python
import dmrg1 as dmrg1
import tdvp
import MPO_class as MPO
import MPS_class as MPS
from ncon import ncon

import matplotlib.pyplot as plt
import numpy as np

plt.rc('text',usetex=True)
```

```python
L   = 64
chi = 64

delta  = -1.2

H = MPO.XXZMPO(L, delta,0.)

engineDMRG = dmrg1.DMRG1(H)
engineDMRG.initialize(chi)

for i in range(20):
    engineDMRG.dmrg_step()
```

```python
# # # COMPUTE <0| s_j^z(t_n) s_{L/2}^z(0) |0> # # #
def compute_corr(MPS0,MPS1,op):
    Mtemp0 = MPS0.M
    Mtemp1 = MPS1.M
    L     = len(Mtemp0)
    corr  = np.zeros(L) 
    for i in range(0,L):
        R = np.ones((1,1,1))
        for j in range(L-1,-1,-1):
            if j == i:
                optemp = op.reshape(1,1,2,2)
            else:
                optemp = np.eye(2).reshape(1,1,2,2)
            R = ncon([R,Mtemp0[j].conj(),Mtemp1[j],optemp],[[1,3,4],[-3,5,4],[-1,2,1],[-2,3,2,5]])   
        corr[i] = (R[0][0][0]).real
    return corr
```

```python
sigmax = np.array([[0,1],[1,0]])

excitedMPS   = MPS.MPS(L,chi,2)
excitedMPS.M = engineDMRG.MPS.M.copy()
excitedMPS.M[L//2] = ncon([excitedMPS.M[L//2],sigmax],[[-1,1,-3],[1,-2]])

T  = 20
dt = 0.1
nsteps = int(T/dt)
Czz = np.zeros((nsteps+1,L))

Cxx[0,:] = compute_corr(engineDMRG.MPS, excitedMPS, sigmax)
```

```python
engineTDVP = tdvp.TDVP(excitedMPS, H)
engineTDVP.initialize()
for n in range(1,nsteps+1):
    engineTDVP.time_step(dt)
    Cxx[n,:] = compute_corr(engineDMRG.MPS,engineTDVP.MPS,sigmax)
```

```python
Cxxcc = np.zeros_like(Cxx,dtype=complex)
for i in range(nsteps):
    Cxxcc[i,:] = np.exp(1j*engineDMRG.E*i*dt)*Cxx[i,:]-Cxx[0,:]
```

```python
import seaborn as sns
import matplotlib

cmap = sns.color_palette('icefire',as_cmap=True)
plt.figure(figsize=(12,7))

plt.imshow(Cxxcc.real,origin='lower',aspect='auto',extent=[0,L,0,nsteps*dt],cmap=cmap,norm=matplotlib.colors.SymLogNorm(linthresh=1e-3,vmin=-1,vmax=1))
plt.tick_params(labelsize=22)

cb = plt.colorbar()
cb.ax.tick_params(labelsize=22)
cb.set_label(r"$\mathrm{Re}\left[S^{xx}_\mathrm{0,L/2}(j,t_n)\right]$",fontsize=28)
plt.ylabel(r'$t$',fontsize=32)
plt.xlabel(r'$j$',fontsize=32)

plt.tight_layout()

plt.savefig('light_cone.jpg',format='jpg',bbox_inches='tight',dpi=800)
```

```python
Sxx = np.fft.fftshift(np.fft.fft2(Cxxcc,norm='ortho'),axes=[0,1])
freq = np.fft.fftshift(np.fft.fftfreq(nsteps,dt))*2*np.pi
```

```python
plt.figure(figsize=(14,10))
cmap1 = sns.color_palette('icefire',as_cmap=True)
plt.imshow(Sxx.real,origin='lower',cmap=cmap1,aspect='auto',extent=[0,2*np.pi,freq[0],freq[-1]],norm=matplotlib.colors.SymLogNorm(linthresh=1e-1,vmin=-2,vmax=2))
lim = np.max(np.abs(Sxx.real))

plt.tick_params(labelsize=22)
cb = plt.colorbar()
cb.ax.tick_params(labelsize=22)
cb.set_label(r"$\mathrm{Re}\left[S^{xx}_\mathrm{0,L/2}(k,\omega)\right]$",fontsize=32)
plt.xlabel("$k$",fontsize=32)
plt.ylabel(r"$\omega$",fontsize=32)
plt.xticks(np.linspace(0,2*np.pi,5),[r'$0$',r'$\pi/2$',r'$\pi$',r'$3\pi/2$',r'$2\pi$'])
plt.savefig('DSSF.png',format='png',dpi=800,bbox_inches='tight')
```
