```python
import dmrg1 as dmrg1
import tdvp
import MPO_class as MPO
import MPS_class as MPS

import matplotlib.pyplot as plt
import numpy as np

plt.rc('text',usetex=True)
```

```python
L = 64
chi = 64

Mz  = MPO.getMzMPO(L)
Mx  = MPO.getMxMPO(L)

h_space = np.linspace(-2,2,128)
mzh = np.zeros_like(h_space)
mxh = np.zeros_like(h_space)
eh  = np.zeros((h_space.size,2))
for i,h in enumerate(h_space):
    H   = MPO.IsingMPO(L,h)
    engine = dmrg1.DMRG1(H)
    engine.initialize(chi)
    for n in range(10):
        engine.dmrg_step()
    mzh[i] = Mz.contractMPOMPS(engine.MPS)
    mxh[i] = Mx.contractMPOMPS(engine.MPS)
    eh[i,0] = engine.E/L
    eh[i,1] = engine.check_convergence()
```

```python
plt.figure(figsize=(15,6))
plt.subplot(121)
plt.plot(h_space,mzh)
plt.tick_params(labelsize=22)
plt.xlabel("$h$",fontsize=30)
plt.ylabel("$m_z$",fontsize=30)
plt.subplot(122)
plt.plot(h_space,mxh)
plt.tick_params(labelsize=22)
plt.xlabel("$h$",fontsize=30)
plt.ylabel("$m_x$",fontsize=30)

plt.tight_layout()
```

```python
plt.figure(figsize=(15,6))
plt.subplot(121)
plt.plot(h_space,eh[:,0])
plt.tick_params(labelsize=22)
plt.xlabel("$h$",fontsize=30)
plt.ylabel("$E/L$",fontsize=30)
plt.subplot(122)
plt.plot(h_space,np.abs(eh[:,1]))
plt.tick_params(labelsize=22)
plt.xlabel("$h$",fontsize=30)
plt.ylabel("$\sigma_E$",fontsize=30)
plt.yscale('log')
plt.tight_layout()
```

```python
L = 128
chi = 256
h = 1

H   = MPO.IsingMPO(L,h)
engine = dmrg1.DMRG1(H)
engine.initialize(chi)
for n in range(10):
    engine.dmrg_step()
```

```python
sigma_x = np.array([[0,1],[1,0]]).reshape(1,1,2,2)

Cxx = np.ones(L-1)
MPStemp = MPS.MPS(L,20,2)
MPStemp.M = engine.MPS.M.copy()
for j in range(1,L):
    Cxx[j-1] = MPO.ComputeCorrFunction(MPStemp,0,j,sigma_x,sigma_x).real
```

```python
from scipy.optimize import curve_fit
def power_law(x,a,b):
    return b*x**a

plt.figure(figsize=(8,6))
plt.plot(Cxx)
plt.tick_params(labelsize=22)

plt.xscale('log')
plt.yscale('log')

popt,pcov = curve_fit(power_law,np.arange(3,50),Cxx[3:50])

plt.plot(np.arange(1,L),power_law(np.arange(1,L),popt[0],popt[1]),label="$%.2f j^{%.2f}$"%(popt[1],popt[0]))
plt.legend(fontsize=18)
plt.xlabel("$j$",fontsize=28)
plt.ylabel(r"$\left\langle \sigma^x_{0}\sigma^x_{j}\right\rangle$",fontsize=28)
```
