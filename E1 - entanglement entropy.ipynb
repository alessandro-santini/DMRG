```python
import dmrg1 as dmrg1
import MPO_class as MPO

import matplotlib.pyplot as plt
import numpy as np

plt.rc('text',usetex=True)
```

```python
# DMRG parameters
nsteps = np.arange(6)
chi = 64

# Model parameters
L   = 128
h   = 1.
H   = MPO.IsingMPO(L,1.)
engine = dmrg1.DMRG1(H)

engine.initialize(chi)
for n in range(10):
    engine.dmrg_step()
```

```python
Sent = engine.MPS.compute_EntEntropy()
```

```python
plt.figure(figsize=(7,5))

plt.plot(np.arange(1,L),Sent)
plt.xticks(np.arange(0,129,16))
plt.tick_params(labelsize=22)
plt.xlabel("$l$",fontsize=22)
plt.ylabel("$S(L,l)$",fontsize=22)
```

```python
from scipy.optimize import curve_fit

def exactfun(x,c,k):
    return c/3*np.log(np.sin(np.pi*x))+k

x_space = np.arange(1,L)/L
popt,pcov = curve_fit(exactfun, x_space,Sent)
```

```python
plt.figure(figsize=(8,6))
loverL = np.linspace(1,L-1,1024)/L
plt.plot(np.linspace(1,L-1,1024), exactfun(loverL,popt[0],popt[1]),'k',linewidth=0.7)
plt.plot(np.arange(1,L),Sent,'o',markersize=4,color='xkcd:crimson')
plt.tick_params(labelsize=20)

plt.xlabel("$l$", fontsize=22)
plt.ylabel("$S(L,l)$", fontsize=22)
```

```python
# DMRG parameters
nsteps = np.arange(6)
chi = 64

# Model parameters
L     = 128
h     = 0.
delta = 0.2
H     = MPO.XXZMPO(L,delta,h)
engine = dmrg1.DMRG1(H)

engine.initialize(chi)
for n in range(10):
    engine.dmrg_step()
```

```python
Sent = engine.MPS.compute_EntEntropy()

x_space = np.arange(1,L)/L
popt,pcov = curve_fit(exactfun, x_space,Sent)

plt.figure(figsize=(8,6))
loverL = np.linspace(1,L-1,1024)/L
plt.plot(np.linspace(1,L-1,1024), exactfun(loverL,popt[0],popt[1]),'k',linewidth=0.7)
plt.plot(np.arange(1,L),Sent,'o',markersize=4,color='xkcd:crimson')
plt.tick_params(labelsize=20)

plt.xlabel("$l$", fontsize=22)
plt.ylabel("$S(L,l)$", fontsize=22)
```
